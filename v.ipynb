```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Set seed for reproducibility
random.seed(42)

# Sample size
num_records = 10000

# Sample data lists
job_titles = [
    "Software Engineer", "Data Scientist", "Product Manager", "UI/UX Designer",
    "DevOps Engineer", "Backend Developer", "Frontend Developer", "Business Analyst",
    "Machine Learning Engineer", "Cybersecurity Analyst"
]

industries = [
    "Technology", "Finance", "Healthcare", "Retail", "Education",
    "Manufacturing", "Telecom", "Logistics", "Media", "Energy"
]

locations = [
    "Bangalore", "Mumbai", "Delhi", "Hyderabad", "Chennai", "Pune",
    "Remote", "Gurgaon", "Noida", "Ahmedabad"
]

company_names = [
    "Infosys", "TCS", "Wipro", "Amazon", "Google", "Microsoft",
    "Flipkart", "Paytm", "Accenture", "Capgemini"
]

education_levels = ["Bachelor's", "Master's", "PhD", "Diploma", "High School"]

experience_levels = ["Fresher", "1-2 years", "3-5 years", "6-10 years", "10+ years"]

salary_ranges = {
    "Fresher": (300000, 600000),
    "1-2 years": (500000, 800000),
    "3-5 years": (800000, 1200000),
    "6-10 years": (1200000, 2000000),
    "10+ years": (2000000, 3500000)
}

job_types = ["Full-time", "Part-time", "Internship", "Contract"]
skills_pool = ["Python", "Java", "SQL", "JavaScript", "Django", "React", "Node.js", "AWS", "Machine Learning", "Data Analysis"]

data = []

for i in range(num_records):
    title = random.choice(job_titles)
    industry = random.choice(industries)
    location = random.choice(locations)
    company = random.choice(company_names)
    education = random.choice(education_levels)
    experience = random.choice(experience_levels)
    salary_min, salary_max = salary_ranges[experience]
    salary = random.randint(salary_min, salary_max)
    remote_option = "Yes" if location == "Remote" else random.choice(["Yes", "No"])
    hired = random.choice(["Yes", "No"])
    job_type = random.choice(job_types)
    skills = ", ".join(random.sample(skills_pool, k=random.randint(2, 5)))
    posted_date = datetime.today() - timedelta(days=random.randint(1, 60))
    deadline_date = posted_date + timedelta(days=random.randint(10, 30))
    interview_rounds = random.randint(1, 5)
    job_description = f"{title} role at {company} requiring skills in {skills.split(',')[0]} and more."
    company_rating = round(random.uniform(2.5, 5.0), 1)
    joining_time_days = random.randint(7, 60)
    stipend = random.randint(5000, 25000) if job_type == "Internship" else 0
    job_id = f"JOB2025{str(i+1).zfill(3)}"

    data.append({
        "Job ID": job_id,
        "Job Title": title,
        "Industry": industry,
        "Location": location,
        "Company": company,
        "Education Level": education,
        "Experience Level": experience,
        "Salary (INR)": salary,
        "Remote Option": remote_option,
        "Hired": hired,
        "Job Type": job_type,
        "Skills Required": skills,
        "Posted Date": posted_date.date(),
        "Application Deadline": deadline_date.date(),
        "Interview Rounds": interview_rounds,
        "Job Description Snippet": job_description,
        "Company Rating": company_rating,
        "Expected Joining Time (days)": joining_time_days,
        "Internship Stipend": stipend
    })

# Create DataFrame and export
df_full = pd.DataFrame(data)
file_path_full = "/mnt/data/job_market_analysis_full_data.csv"
df_full.to_csv(r'C:\Users\gupta\Desktop\Hack and viz project 1\job_market_analysis_full_data.csv', index=False)

file_path_full
```
